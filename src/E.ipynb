# Analiza Supraviețuirii pe Titanic
Acest notebook prezintă o analiză exploratorie a setului de date **Titanic**, urmărind influența sexului și a clasei sociale asupra ratei de supraviețuire.

```python
# ==========================================
# 1. Importul bibliotecilor și încărcarea datelor
# ==========================================
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Setăm tema vizuală
sns.set_theme(style='whitegrid', palette='muted')

# Încărcăm datele
titanic = sns.load_dataset('titanic')
print("Dataset încărcat cu succes!")
```

## Explorarea Datelor
Vom verifica structura tabelului și vom identifica valorile lipsă.

```python
# Afișăm primele rânduri și info
print(f"Dimensiune: {titanic.shape}")
print(f"\nRata globală de supraviețuire: {titanic['survived'].mean():.1%}")

# Verificăm valorile lipsă
missing = titanic.isnull().sum()
print(f"\nValori lipsă:\n{missing[missing > 0]}")
```

## Vizualizarea Rezultatelor
Vom genera grafice pentru a compara ratele de supraviețuire în funcție de clasă și sex.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Subplot 1: Supraviețuire per sex
sns.barplot(data=titanic, x='sex', y='survived', hue='sex', ax=axes[0])
axes[0].set_title('Supraviețuire după Sex')

# Subplot 2: Supraviețuire per clasă și sex
sns.barplot(data=titanic, x='class', y='survived', hue='sex', ax=axes[1])
axes[1].set_title('Supraviețuire după Clasă și Sex')

plt.tight_layout()
# În Colab, graficul apare automat aici.
```

## Concluzii
1. **Genul a fost determinant:** Femeile au avut o rată de supraviețuire mult mai mare decât bărbații în toate clasele.
2. **Clasa socială a contat:** Pasagerii de la clasa I au avut prioritate sau acces mai facil la bărcile de salvare față de cei de la clasa a III-a.
3. **Efect combinat:** O femeie de la clasa I a avut cele mai mari șanse de supraviețuire, în timp ce un bărbat de la clasa a III-a a avut cele mai mici șanse.
